## <font color=RoyalBlue>Logistic Regression<font>

Logistic regression is a type of regression analysis in statistics used for prediction of outcome of a categorical dependent variable from a set of predictor or independent variables. In logistic regression the dependent variable is always binary. Logistic regression is mainly used to for prediction and also calculating the probability of success.

```python
from statsmodels.tools import add_constant as add_constant
# add intercept term in heart_df
heart_df_constant = add_constant(heart_df)
heart_df_constant.head()
```

```python
# input feature cols
cols=heart_df_constant.columns[:-1]
cols
```

```python
model=sm.Logit(heart_df.TenYearCHD,heart_df_constant[cols])
result=model.fit()
result.summary()
```

The results above show some of the attributes with P value higher than the preferred alpha(5%) and thereby showing  low statistically significant relationship with the probability of heart disease. Backward elemination approach is used here to remove those attributes with highest Pvalue one at a time follwed by  running the regression repeatedly until all attributes have P Values less than 0.05.



### <font color=CornflowerBlue>Feature Selection: Backward elemination (P-value approach)<font>

```python
def back_feature_elem (data_frame,dep_var,col_list):
    """ Takes in the dataframe, the dependent variable and a list of column names, runs the regression repeatedly eleminating feature with the highest
    P-value above alpha one at a time and returns the regression summary with all p-values below alpha"""

    while len(col_list)>0 :
        model=sm.Logit(dep_var,data_frame[col_list])
        result=model.fit(disp=0)
        largest_pvalue=round(result.pvalues,3).nlargest(1) # select single largest p-value
        if largest_pvalue[0]<(0.05):
            return result
            break
        else:
            col_list=col_list.drop(largest_pvalue.index)

result=back_feature_elem(heart_df_constant,heart_df.TenYearCHD,cols)
```

```python
result.summary()
```

#### <font color=darkblue>Logistic regression equation<font>

$$P=\hspace{.2cm}e^{\beta_0 + \beta_1 X_1}\hspace{.2cm}/\hspace{.2cm}1+e^{\beta_0 +\beta_1 X_1}$$

When all features plugged in:

$$logit(p) = log(p/(1-p))=\beta_0 +\beta_1\hspace{.1cm} *\hspace{.2cm} Sexmale\hspace{.2cm}+\beta_2\hspace{.1cm} * \hspace{.1cm}age\hspace{.2cm}+\hspace{.2cm}\beta_3\hspace{.1cm} *\hspace{.1cm} cigsPerDay\hspace{.2cm}+\hspace{.2cm}\beta_4 \hspace{.1cm}*\hspace{.1cm} totChol\hspace{.2cm}+\hspace{.2cm}\beta_5\hspace{.1cm} *\hspace{.1cm} sysBP\hspace{.2cm}+\hspace{.2cm}\beta_6\hspace{.1cm} *\hspace{.1cm} glucose\hspace{.2cm}$$

### <font color=CornflowerBlue>Splitting data to train and test split<font>

```python
import sklearn
new_features=heart_df[['age','Sex_male','cigsPerDay','totChol','sysBP','glucose','TenYearCHD']]
x=new_features.iloc[:,:-1]
y=new_features.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=5)

# Train - 80%
# Test - 20%
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
```

### <font color=CornflowerBlue>Model Evaluation - Statistics<font>

```python
print('The acuuracy of the model = TP+TN/(TP+TN+FP+FN) = ',(TP+TN)/float(TP+TN+FP+FN),'\n',

'The Missclassification = 1-Accuracy = ',1-((TP+TN)/float(TP+TN+FP+FN)),'\n',

'Sensitivity or True Positive Rate = TP/(TP+FN) = ',TP/float(TP+FN),'\n',

'Specificity or True Negative Rate = TN/(TN+FP) = ',TN/float(TN+FP),'\n',

'Positive Predictive value = TP/(TP+FP) = ',TP/float(TP+FP),'\n',

'Negative predictive Value = TN/(TN+FN) = ',TN/float(TN+FN),'\n',

'Positive Likelihood Ratio = Sensitivity/(1-Specificity) = ',sensitivity/(1-specificity),'\n',

'Negative likelihood Ratio = (1-Sensitivity)/Specificity = ',(1-sensitivity)/specificity)
```

Sure! Let’s break each of these terms in simple **layman terms** using your example of **detecting heart disease (Yes/No)**.

---

### 1️⃣ **Positive Predictive Value (PPV)**

**Formula:**

$$
PPV = \frac{TP}{TP + FP}
$$

* **TP (True Positive):** Number of people correctly predicted as having heart disease.
* **FP (False Positive):** Number of people incorrectly predicted as having heart disease (they are actually healthy).

**What it means:**

* If your test says “Yes, you have heart disease,” **PPV tells you how likely it is actually true.**
* Example: PPV = 0.8 → If 10 people test positive, 8 actually have heart disease.

---

### 2️⃣ **Negative Predictive Value (NPV)**

**Formula:**

$$
NPV = \frac{TN}{TN + FN}
$$

* **TN (True Negative):** Number of people correctly predicted as **not having heart disease**.
* **FN (False Negative):** Number of people incorrectly predicted as **not having heart disease** (they actually have it).

**What it means:**

* If your test says “No, you don’t have heart disease,” **NPV tells you how likely that is true.**
* Example: NPV = 0.9 → If 10 people test negative, 9 actually don’t have heart disease.

---

### 3️⃣ **Positive Likelihood Ratio (PLR)**

**Formula:**

$$
PLR = \frac{Sensitivity}{1 - Specificity}
$$

* **Sensitivity:** How good the test is at finding people who **do have heart disease**.
* **Specificity:** How good the test is at correctly identifying people who **don’t have heart disease**.

**What it means:**

* PLR tells you **how much more likely a positive test result is in someone with heart disease compared to someone without it.**
* Higher PLR → The test is very good at confirming disease when it’s present.

---

### 4️⃣ **Negative Likelihood Ratio (NLR)**

**Formula:**

$$
NLR = \frac{1 - Sensitivity}{Specificity}
$$

* Measures **how likely a negative test result is in someone with the disease compared to someone without it.**

**What it means:**

* Lower NLR → The test is very good at ruling out heart disease when the test is negative.

---

✅ **In short:**

| Term | Simple Meaning                                        |
| ---- | ----------------------------------------------------- |
| PPV  | If test says “Yes,” how likely is it really yes?      |
| NPV  | If test says “No,” how likely is it really no?        |
| PLR  | How strongly does a positive test suggest disease?    |
| NLR  | How strongly does a negative test suggest no disease? |

### <font color=CornflowerBlue>Lower the threshold<font>

Since the model is predicting Heart disease too many type II errors is not advisable. A False Negative ( ignoring the probability of disease when there actualy is one) is more dangerous than a False Positive in this case. Hence inorder to increase the sensitivity,  threshold can be lowered.

```python
# FN is important to us in this case, lower the threshold (to get yes -> no )
# FP is very important increase threshold (no -> yes)
from sklearn.preprocessing import binarize
for i in range(1,5):
    cm2=0
    y_pred_prob_yes=logreg.predict_proba(x_test)
    y_pred2=binarize(y_pred_prob_yes, threshold=i/10)[:,1] # Use keyword argument for threshold
    cm2=confusion_matrix(y_test,y_pred2)
    print ('With',i/10,'threshold the Confusion Matrix is ','\n',cm2,'\n',
            'with',cm2[0,0]+cm2[1,1],'correct predictions and',cm2[1,0],'Type II errors( False Negatives)','\n\n',
          'Sensitivity: ',cm2[1,1]/(float(cm2[1,1]+cm2[1,0])),'Specificity: ',cm2[0,0]/(float(cm2[0,0]+cm2[0,1])),'\n\n\n')
```

### <font color=CornflowerBlue>ROC curve<font>

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob_yes[:,1])
plt.plot(fpr,tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for Heart disease classifier')
plt.xlabel('False positive rate (1-Specificity)')
plt.ylabel('True positive rate (Sensitivity)')
plt.grid(True)
```

In an ROC curve like this:

Bottom-left (0,0) → This corresponds to a high decision threshold (close to 1).

The classifier predicts positive only when it’s very sure.

Both TPR (True Positive Rate) and FPR (False Positive Rate) are low, meaning you detect few positives but also make very few false alarms.

Top-right (1,1) → This corresponds to a low decision threshold (close to 0).

The classifier predicts positive for almost everyone.

Both TPR and FPR are high — you catch almost all positives but also have many false alarms.

A common way to visualize the trade-offs of different thresholds is by using an ROC curve, a plot of the true positive rate (# true positives/ total # positives) versus the false positive rate (# false positives /
total # negatives) for all possible choices of thresholds. A model with good classification accuracy should have significantly more true positives than false positives at all thresholds.

The optimum position for roc curve is towards the top left corner where the specificity and sensitivity are at optimum levels

Probability interpretation: If you randomly pick one patient with heart disease and one without, your model will give the patient with heart disease a higher risk score about 75–80% of the time.

### <font color=CornflowerBlue>Area Under The Curve (AUC)<font>

The area under the ROC curve quantifies model classification accuracy; the higher the area, the greater the disparity between true and false positives, and the stronger the model in classifying members of the training dataset. An area of 0.5 corresponds to a model that performs no better than random classification and a good classifier stays as far away from that  as possible. An area of 1 is ideal. The closer the AUC to 1 the better.
